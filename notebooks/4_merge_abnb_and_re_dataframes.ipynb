```python
import numpy as np
import pandas as pd
from google.colab import drive
```

```python
drive = drive.mount('/content/drive/')
data_path = '/content/drive/My Drive/Colab_Data/AirBnb/Italy/Sicily/'

#save the DF so we don't have to keep doing geo lookups
file_path = data_path + f'derived/abnb_geo_df_500.csv'
abnb_df = pd.read_csv(file_path)
```

```python
abnb_df.drop(columns=['Unnamed: 0', 'Unnamed: 0.1'], inplace=True)
abnb_df.head()
```

```python
re_file_path = data_path + 'derived/re_municipalities.csv'
re_df = pd.read_csv(re_file_path)
re_df.head()
```

```python
abnb_df[['sale_low', 'sale_mean', 'sale_high']] = np.nan

for index, row in abnb_df.iterrows():
  # print(f"geo_common_name: {row['geo_common_name']}")
  re_match = re_df[re_df['city'] == row['geo_common_name']]
  # print(f"geo_common_name: {row['geo_common_name']}::{len(re_match)}")

  if len(re_match) == 1:
    # print(f"1 MATCH: geo_common_name: {row['geo_common_name']}::{len(re_match)}")  
    
    abnb_df['sale_low'][index] = re_match['sale_low']
    abnb_df['sale_mean'][index] = re_match['sale_mean']
    abnb_df['sale_high'][index] = re_match['sale_high']

  elif len(re_match) > 1:
    print(f"> 1 MATCH: geo_common_name: {row['geo_common_name']}::{len(re_match)}")
  else:
    print(f"NO MATCH: geo_common_name: {row['geo_common_name']}::{len(re_match)}")

abnb_df.head()
```

```python
abnb_df.dropna(subset=['geo_common_name', 'sale_low', 'sale_mean', 'sale_high'], inplace=True)
abnb_df.shape
```

```python
#save the DF so we don't have to keep doing geo lookups
file_path = data_path + f'derived/abnb_geo_with_re_prices_df_{len(abnb_df)}.csv'
abnb_df.to_csv(file_path)
```
